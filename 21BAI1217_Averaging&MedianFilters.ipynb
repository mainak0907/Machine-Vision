Averaging Low Pass Filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image
```

```python
# Read the image
img = cv2.imread('./images.jpg', 0)
assert not isinstance(img,type(None)), 'image not found'
plt.imshow(img)
```

```python
# Obtain number of rows and columns
# of the image
m, n = img.shape
```

```python
# Develop Averaging filter(3, 5) mask
mask = np.ones([3, 3], dtype = int)
mask = mask / 9
```

```python
# Convolve the 3X3 mask over the image
img_new = np.zeros([m, n])

for i in range(1, m-1):
    for j in range(1, n-1):
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2]

        img_new[i, j]= temp

img_new = img_new.astype(np.uint8)
plt.imshow(img_new)
cv2.imwrite('blurred.tif', img_new)
```

```python
# Develop Averaging filter (5, 5) mask
mask = np.ones([5, 5], dtype=int)
mask = mask / 25  # Normalize the mask
```

```python
# Convolve the 5x5 mask over the image
img_new = np.zeros([m, n])
```

```python
for i in range(2, m-2):
    for j in range(2, n-2):
        temp = (
            img[i-2, j-2]*mask[0, 0] + img[i-2, j-1]*mask[0, 1] + img[i-2, j]*mask[0, 2] + img[i-2, j+1]*mask[0, 3] + img[i-2, j+2]*mask[0, 4] +
            img[i-1, j-2]*mask[1, 0] + img[i-1, j-1]*mask[1, 1] + img[i-1, j]*mask[1, 2] + img[i-1, j+1]*mask[1, 3] + img[i-1, j+2]*mask[1, 4] +
            img[i, j-2]*mask[2, 0] + img[i, j-1]*mask[2, 1] + img[i, j]*mask[2, 2] + img[i, j+1]*mask[2, 3] + img[i, j+2]*mask[2, 4] +
            img[i+1, j-2]*mask[3, 0] + img[i+1, j-1]*mask[3, 1] + img[i+1, j]*mask[3, 2] + img[i+1, j+1]*mask[3, 3] + img[i+1, j+2]*mask[3, 4] +
            img[i+2, j-2]*mask[4, 0] + img[i+2, j-1]*mask[4, 1] + img[i+2, j]*mask[4, 2] + img[i+2, j+1]*mask[4, 3] + img[i+2, j+2]*mask[4, 4]
        )

        img_new[i, j] = temp
```

```python
img_new = img_new.astype(np.uint8)
cv2.imwrite('new2.png', img_new)
```

Median Filter

```python
# Median Spatial Domain Filtering
import cv2
import numpy as np
```

```python
# Read the image
img_noisy1 = cv2.imread('./download (1).jpeg', 0)
assert not isinstance(img_noisy1,type(None)), 'image not found'
plt.imshow(img_noisy1)
```

```python
# Obtain the number of rows and columns
# of the image
m, n = img_noisy1.shape
```

```python
# Traverse the image. For every 3X3 area,
# find the median of the pixels and
# replace the center pixel by the median
img_new1 = np.zeros([m, n])

for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img_noisy1[i-1, j-1],
               img_noisy1[i-1, j],
               img_noisy1[i-1, j + 1],
               img_noisy1[i, j-1],
               img_noisy1[i, j],
               img_noisy1[i, j + 1],
               img_noisy1[i + 1, j-1],
               img_noisy1[i + 1, j],
               img_noisy1[i + 1, j + 1]]

        temp = sorted(temp)
        img_new1[i, j]= temp[4]

img_new1 = img_new1.astype(np.uint8)
plt.imshow(img_new1)
cv2.imwrite('new_median_filtered.png', img_new1)
```
