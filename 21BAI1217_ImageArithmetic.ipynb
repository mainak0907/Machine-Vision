```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

Image Arithmetic

```python
def resize_images(img1, img2):
    # Resize images to have the same shape
    min_height = min(img1.shape[0], img2.shape[0])
    min_width = min(img1.shape[1], img2.shape[1])

    resized_img1 = np.array(Image.fromarray(img1).resize((min_width, min_height)))
    resized_img2 = np.array(Image.fromarray(img2).resize((min_width, min_height)))

    return resized_img1, resized_img2
```

```python
def image_addition(img1, img2):
    # Ensure images have the same shape
    assert img1.shape == img2.shape, "Images must have the same shape for addition."

    # Perform addition
    result_img = img1 + img2

    # Clip values to ensure they are within the valid range [0, 255]
    result_img = np.clip(result_img, 0, 255)

    return result_img.astype(np.uint8)
```

```python
def image_subtraction(img1, img2):
    # Ensure images have the same shape
    assert img1.shape == img2.shape, "Images must have the same shape for subtraction."

    # Perform subtraction
    result_img = img1 - img2

    # Clip values to ensure they are within the valid range [0, 255]
    result_img = np.clip(result_img, 0, 255)

    return result_img.astype(np.uint8)
```

```python
img1 = plt.imread('image1.jpg')
img2 = plt.imread('image2.jpg')
```

```python
# Resize images
img1, img2 = resize_images(img1, img2)
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
result_addition = image_addition(img1, img2)
result_subtraction = image_subtraction(img1, img2)
```

```python
plt.imshow(result_addition)
plt.title("Addition Result")
plt.show()
```

```python
plt.imshow(result_subtraction)
plt.title("Subtraction Result")
plt.show()
```
